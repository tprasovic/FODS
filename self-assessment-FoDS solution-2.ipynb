Given the list `a = [el for el in range(-5, 5, 1)]`

### 3.1 -  while loop
Using a **while** loop compute the factorial of the number 5.\
Factorial is a mathematical operation denoted by the symbol **"!"** that multiplies a given number by all positive integers less than itself. For example, the factorial of a positive integer *n*, denoted as *n!*, is the product of all positive integers from 1 up to *n*.

$5! = 5*4*3*2*1$


````python
num = 5
# Initialize variables
factorial = num
counter = factorial - 1

# Calculate factorial using a while loop
while counter >= 1:
    factorial *= counter
    counter -= 1

# Print the factorial
print(f"The factorial of {num} is {factorial}")
````

## 4 - Python packages and functions

In this part, you are asked to import the `numpy` library to perform some mathematical operations.
- The task is to write a Python function (`my_average`) to compute the average of a numpy array without using numpy, and then check using Python's built-in `assert` that the result of the function and that of the built-in numpy method match. The function should take as input an array `a`**` and return the average of the array.

````python
##### import here
import numpy as np


### create array with integer numbers from 1 to 20
a = np.arange(1, 20)

### function here
def my_average(array):
    my_sum = 0
    for num in array:
        my_sum += num
    return my_sum / len(array)

if np.mean(a) == my_average(a):
    print('The results match!')
else:
    assert np.mean(a) == my_average(a), 'The results are different. Check your function.'
````

# 1. Basics of Data visualization with python

**Definition**: The art of communicating meaningful data visually. This can involve infographics, traditional plots, or even full data dashboards. Some tips for data visualization: 

* Choose a consistent  graph type for each data type
* Always include plot title, axis titles, and units. Include a legend when applicable. 
* Use appropriate colors and color palettes. Think about how they will affect a color blind audience. Divergent colormaps can be useful to enhance differences, whereas sequential colormaps are more appropriate for progression data.

How to create your graph frame and multiple subplots using `plt.subplots` 

```python
# Create a Figure and axes with plt.subplots with figure size (12,4)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12,4))

#ax1 and ax2 are an empty set of axes
plt.show()
```

Fill in the subplots using `plot` and `scatter`. 

```python
import numpy as np

# create a vector to use for the x-axis
x = np.linspace(0, 2*np.pi, 100)
y =  np.sin(x)

# you can plot a line 
line1, = ax1.plot(x, y, label="sin(x)")

# create random noise points to sum to the function
z = y + 0.5 * np.random.rand(len(x))

# create the representation as scatter points
p1= ax2.scatter(x, z, marker=".")

# in jupyter, you can also just call the figure to show the result
fig
```

```python
# generate cosine functions 
y2 = np.cos(x)
z2 = y2 + 0.2 * np.random.rand(len(x))

# plot another customize style line still using the plot created earlier
line2, = ax1.plot(x, y2,  color="red", linewidth=3, label="cos(x)")
p2 = ax2.scatter(x, z2, marker="x", s=15, color="red",linewidth=1)
fig
```

How to set `axis limits`, `title`, `labels`, `legends`, use `ticks` and customise your plots. 

```python
# set axis limits
ax1.set_xlim(0, 2* np.pi)
ax1.set_ylim(-1.02, 1.02)

# set plot title and axes labels
ax1.set(title = "Trigonometric Functions",  xlabel = "angle [radian]",  ylabel = "f(x)");

# customize the legend to add a legend title and place it at a manually specified location 
ax1.legend(title="f(x)",handles=[line1, line2], loc="upper center", bbox_to_anchor=(0.6, 1.0))

# add semitrasparent dotted (-) grid lines
ax1.grid(color="gray", alpha=0.3, linestyle=":", linewidth=1)

# pad margins so that markers don"t get clipped by the axes
ax1.margins(0.2)

# set plot title and axes labels
ax2.set(title = "Noisy Data Points", xlabel = "angle [radian]", ylabel = "g[x]")

# set minor ticks for more precission 
ax2.minorticks_on()

# we can customize the tick labels, e.g. using multiples of PI
x_tick = np.arange(0, 2.1*np.pi, 0.5*np.pi) 
# use LaTeX to format strings nicely if they contain mathematical symbols
x_text = [r"$0$", r"$\frac{\pi}{2}$",   r"$\pi$", r"$\frac{3\pi}{2}$", r"${2\pi}$",] 
# set text labels and rotate them if appropiate. 
ax2.set_xticks(x_tick, x_text, rotation=0)  

# sweak spacing to prevent clipping of tick-labels
plt.subplots_adjust(bottom=0.15)

# Customize the legend with new labels
ax2.legend([p1, p2], ["sin[x] + n", "cos[x]+ n"], loc="best")

fig
```

## Colors: color scales and palettes 

Use color only when needed to serve a particular communication goal. Use different colors only when they correspond to differences of meaning in the data.

- ***qualitative color scale***:  use color as a means to distinguish discrete items or groups that do not have an intrinsic order such as different countries
- ***sequential color scale***: represent data values such as temperature range
- ***accent color scales***: use color to highlight. Use soft, natural colors to display most information and bright and/or dark colors to highlight information that requires greater attention. 

In python visualization base library, there are some predefined colors: [[colormap reference]](https://matplotlib.org/stable/tutorials/colors/colormaps.html)

```python
def hex_rgb_norm(hex: str): 
    """ Auxiliary function that takes color representation as a HEX string as 
    input and converts it to a normalized RGB tuple.
        
        Parameters
        ----------
        hex : str
           HEX color string 
           
        Returns
        -------
        tuple 
            normalized [0,1] RGB tuple
    """
    rgb_norm = []
    # Get the hex values without # 
    h = hex.lstrip("#")
    for c in (0, 2, 4):
        # Convert HEX to R,G,B
        color_channel = int(h[c:c+2], 16)
        # Normalize the color
        channel_normalized = color_channel / 255
        # Store the r,g,b channels in a list
        rgb_norm.append(channel_normalized)
    return tuple(rgb_norm) # Convert to tuple for matplotlib compitability
```

# 2. What makes a bad graph? 
## *Ugly, bad, and wrong figures*

Graphs are only as good as the data they display. In general, a good graph would  meet these criteria:

- [x] Easy to communicate. Does not include too many variables
- [x] Enhancing the message. Use consistent scales
- [x] Earn attention and easy to remember

Remember that what makeas a bad plot to be avoided are: 

1. **Ambiguity**: No clear meaning and labelling of what is represented. 
2. **Distortion**: Misleading representation
3. **Distraction**: Visual information that distracts from the visualizations point.

s<img src="https://clauswilke.com/dataviz/introduction_files/figure-html/ugly-bad-wrong-examples-1.png" width="500" align="center" />

Further, a bad plot can be categorized [[source]](https://clauswilke.com/dataviz/introduction.html) as: 

- **ugly**A figure that has aesthetic problems but otherwise is clear and informative.
- **bad**A figure that has problems related to perception; it may be unclear, confusing, overly complicated, or deceiving.
- **wrong**A figure that it is objectively incorrect.

```python
# Sample data to create a bad plot
hospital_disposition = ["ADMITTED","HOME", "TRANSFER", "LEFT AGAINST MEDICAL ADVICE", "LEFT WITHOUT BEING SEEN", "ELOPED"]
patients = [150,  60,   5,   2,   2,  1]

# Create a simple 'bad' plot
fig, axes = plt.subplots(2,1,figsize=(6, 10))
axes[0].plot(hospital_disposition, patients)
axes[0].set_title("Bad Plot Example")

# Create a "better" plot
axes[1].barh(hospital_disposition, patients, color="teal")
plt.gca().invert_yaxis()
plt.title("Hospital Disposition")
plt.ylabel("Disposition")
plt.xlabel("Number of Patients")

plt.show()
```

A `histogram` is a plot that lets you discover and show, the underlying frequency distribution  of a set of continuous data. 

*What other graph types could have been used to visualize age?* 

```python
fig_age, axes = plt.subplots(1, 2, figsize=(10,5))

# For the rest of visualizations with seaborn, we will use the predefined colorblind palette
colorblind_palete = sns.color_palette("colorblind") 

fig_age = sns.histplot(
    data, 
    x = "age", 
    ax=axes[0],
    kde=True
)

label_axes = fig_age.set(
    xlabel="age [years]", 
    ylabel="Number of Patients [count]", 
    title="Age distribution"
) 

# decide on the ranges for the age histogram
bins=[30,35,40,45,50,55,60,65,70,75]

# Plot the hisogram using the gender for color coding
fig_age_custom_bins = sns.histplot(
    data, 
    x = "age", 
    stat="count",
    binwidth=5, 
    binrange=(30, 75),
    hue= "sex", # color by gender 
    kde=True, 
    palette = colorblind_palete, 
    ax=axes[1], 
    multiple="dodge"
)

# Customize the figure labels
fig_age_custom_bins.set(
    xticks=bins, # use bin edges as ticks to make interpreting plot a little easier
    xlabel="age [years]", 
    ylabel="Number of Patients [count]", 
    title= "Age distribution by gender"
)
fig_age_custom_bins.set_xticks( # rotate tick labels so that numbers don't overlap
    fig_age_custom_bins.get_xticks(), 
    fig_age_custom_bins.get_xticklabels(), 
    rotation=90, 
    ha="center"
)

# Change the legend to a more readable format
l = fig_age_custom_bins.legend(title="Gender", labels=["Female", "Male"])
```

`Boxplots` can be easily visualized with python. The corresponding seaborn function will automatically compute:  

- the *median* (middle line in the box)
- *inter-quartile range* (IQR, top and bottom parts of the boxes - this is where 50% of your data is)
- *whiskers* (the black lines extending to the lowest and highest values that are still within 1.5*IQR)
- *outliers* (any observations out with the whiskers)

`Violin plots` is another way to visualise continues variables. 

*What do you observe in the following boxplots an violin plots of serum cholesterol levels and blood pressure (resting)?*

```python
# set the columns we want to plot
columns = ["chol", "trestbps"]
units = ["[mg/dL]", "[mmHg]"]

# Create subplots using a for loop
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

for i, (column, unit) in enumerate(zip(columns*2, units*2)):

    row = i // 2
    col = i % 2

    # Boxplots
    if row == 0:
        sns.boxplot(
            x=data[column], 
            y=data["sex"], 
            order=["female", "male"], 
            hue=data["fbs"],
            ax=axes[row, col]
        )
        # Set the title and labels 
        axes[row, col].set(
            title=f"Boxplots for {column} \n by gender and fasting blood sugar > 120 mg/dl",
            xlabel=f"{column} {unit}", 
            ylabel="Gender"
        )

    #Violin plots
    else:
        sns.violinplot(
            x=data[column], 
            y=data["sex"], 
            order=["female", "male"], 
            hue=data["fbs"],
            ax=axes[row, col]
        )

        # Set the title and labels 
        axes[row, col].set(
            title= f"Violin plots for {column} \n by gender and fasting blood sugar > 120 mg/dl",
            xlabel=f"{column} {unit}", 
            ylabel="Gender"
        )
    

# Adjust layout and show the figure
plt.tight_layout()
plt.show()
```

Correlation  between Biomarkers

`Pair plot` is used to understand the best set of features to explain a relationship between two variables or to form the most separated clusters

```python
# Auxiliary function to customize the label axes
def change_subplots_axes(fig:sns.axisgrid , axes_dict: dict ):
    # iterate over the subplots axes to customize the labels
    for i in range(len(axes_dict)):
        for j in range(i+1):
            xlabel = fig.axes[i][j].get_xlabel()
            ylabel = fig.axes[i][j].get_ylabel()
            if xlabel in axes_dict.keys():
                fig.axes[i][j].set_xlabel(axes_dict[xlabel])
            if ylabel in axes_dict.keys():
                fig.axes[i][j].set_ylabel(axes_dict[ylabel])
```

```python
# Create custom labels
biomarkers_labels = {"chol": "Cholesterol [mg/dL]",
                    "trestbps": "BP (resting) [mm Hg]", 
                    "age": "Age [years]"}

# Define the variables that will be plotted
biomarkers = data[["chol","trestbps", "age", "num"]]

# Plot Pairwise relationships between comorbidity diseases and dead 
fig_correlation = sns.pairplot(biomarkers, 
                               hue ="num", 
                               palette="colorblind"
                            )

# Create labels
change_subplots_axes(fig_correlation, biomarkers_labels)

# # Create interpretable legends
# fig_correlation._legend.remove()
# plt.legend(title="Target Outcome" ,loc="best", bbox_to_anchor=(1.8, 0.5), labels=["heart disease", "healthy"])

plt.suptitle("Pairwise correlation of potential variables", y=0.0)
plt.show()
```
